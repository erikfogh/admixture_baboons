```python
paf_ref_df = pd.read_csv("/home/eriks/XYprimates/people/eriks/sv_papio/results/XY_ref_hamadryas.paf", sep="\t",
                    names=names)
paf_ref_df["identity"] = paf_ref_df.matches/paf_ref_df.bases
paf_ref_df["segment_length"] = paf_ref_df["query_end"] - paf_ref_df["query_start"]
paf_ref_df["tsegment_length"] = paf_ref_df["target_end"] - paf_ref_df["target_start"]
```

```python
paf_ref_df.loc[paf_ref_df.target == "CM001511.2"].bases.sum()
```

```python
paf_ref_df.loc[paf_ref_df.target != "CM001511.2"].matches.sum()
```

```python
chr_list = list(paf_ref_df.groupby(by=["target"])["target_length"].mean().sort_values()[-21:].reset_index().target)
#List of chromosomes, including chrX (CM001511.2)
```

```python
paf_ref_df["match_type"] = ["chrX" if x == "CM001511.2" else "autosome" if x in chr_list else "small_contig" for x in paf_ref_df.target]
```

```python
paf_ref_df["align_total"] = paf_ref_df["query"].map(paf_ref_df.groupby(["query"])["segment_length"].sum())
```

```python
total_segment_matches = paf_ref_df.groupby(["query", "match_type"])["segment_length"].sum().reset_index()
total_segment_matches["align_total"] = total_segment_matches["query"].map(paf_ref_df.groupby(["query"])["segment_length"].sum())
total_segment_matches["percent_aligned"] = total_segment_matches["segment_length"]/total_segment_matches.align_total
```

```python
chrX_list_df = total_segment_matches.loc[(total_segment_matches.match_type == "chrX") & (total_segment_matches.percent_aligned > 0.8)]
chrX_matches_df = paf_ref_df.loc[(paf_ref_df["query"].isin(chrX_list_df["query"])) &
                                 (paf_ref_df.target == "CM001511.2") &
                                 (paf_ref_df.matches > 1000)]
```

```python
chrX_matches_df["q_t_diff_norm"] = (chrX_matches_df["segment_length"]-chrX_matches_df["tsegment_length"])/chrX_matches_df["segment_length"]
```

```python
chrX_matches_df.q_t_diff_norm.sum()
```

```python
sns.scatterplot(data=chrX_matches_df.loc[chrX_matches_df.matches > 100000], x="target_start", y="q_t_diff_norm",
                hue="matches")
```
