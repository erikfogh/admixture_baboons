All SMC

```python
aut_list = glob("../smcpp_pyrho/plots/autosome*.csv")
df_list = []
for aut in aut_list:
    aut_csv = pd.read_csv(aut)
    df_list.append(aut_csv)
aut_dfs = pd.concat(df_list)
```

```python
subset_csv = aut_dfs.loc[(aut_dfs.x > 1000) & (aut_dfs.x < 10000000)]
sns.lineplot(data=subset_csv, x="x", y="y")
plt.xscale('log')
```

```python
chrX_list = glob("../smcpp_pyrho/plots/chrX*.csv")
df_list = []
for X in chrX_list:
    X_csv = pd.read_csv(X)
    df_list.append(X_csv)
x_dfs = pd.concat(df_list)
```

```python
subset_csv = x_dfs.loc[(x_dfs.x > 1000) & (x_dfs.x < 10000000)]
sns.lineplot(data=subset_csv, x="x", y="y")
plt.xscale('log')
```

Combining the dfs

```python
x_dfs["chrom"] = "chrX"
aut_dfs["chrom"] = "autosomes"
concat_dfs = pd.concat([x_dfs, aut_dfs])
subset_csv = concat_dfs.loc[(concat_dfs.x > 2000) & (concat_dfs.x < 10000000)]
sns.lineplot(data=subset_csv, x="x", y="y", hue="chrom")
plt.xscale('log')
```
